```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
new_dataset = df.loc[:, ['Age','Fare','Survived']]
```

```python
new_dataset
```

```python
new_dataset['Age']=new_dataset['Age'].fillna(new_dataset['Age'].mean())
new_dataset.info()
```

```python
new_dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_dataset.drop('Survived', axis=1),
                                                    new_dataset['Survived'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

In this case, scaling actaully helps a little bit as the accuracy score is slighlty higher than that of the actual df.
